###Los roles en Tecnología y Educación presentan deserción más frecuente.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Supongamos que tu DataFrame se llama 'df' y tiene las columnas 'Job Role' y 'Attrition'

# Verificar datos nulos y eliminarlos si es necesario
df = df.dropna(subset=['Job Role', 'Attrition'])

# Convertir 'Attrition' a valores numéricos si está codificado como texto
df['Attrition'] = df['Attrition'].apply(lambda x: 1 if x == 'Left' else 0)

# Filtrar datos para roles en Tecnología y Educación
filtered_df = df[df['Job Role'].isin(['Technology', 'Education'])]

# Calcular la tasa de deserción para los roles filtrados
attrition_by_role = filtered_df.groupby('Job Role')['Attrition'].mean()

# Obtener roles y tasas de deserción como listas
job_roles = attrition_by_role.index.tolist()
attrition_rates = attrition_by_role.values.tolist()

# Crear el gráfico de barras
plt.figure(figsize=(8, 6))
sns.barplot(x=job_roles, y=attrition_rates)  # Utilizar sns.barplot para el gráfico de barras
plt.title('Tasa de Deserción para Roles en Tecnología y Educación')
plt.ylabel('Tasa de Deserción')
plt.xlabel('Rol Laboral')
plt.xticks(rotation=45)  # Rotar etiquetas del eje x para mejor legibilidad
plt.grid(True)  # Agregar una cuadrícula para mejor visualización
plt.show()
```

###Los empleados con ingresos mensuales más bajos tienden a dejar la empresa:


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Supongamos que tu DataFrame se llama 'df' y tiene las columnas 'Monthly Income' y 'Attrition'

# Reemplazar caracteres especiales en los datos
df['Education Level'] = df['Education Level'].str.replace('â€™', '')

# Verificar datos nulos y eliminarlos si es necesario
df = df.dropna(subset=['Monthly Income', 'Attrition'])

# Convertir 'Monthly Income' a valores numéricos si es necesario
df['Monthly Income'] = pd.to_numeric(df['Monthly Income'], errors='coerce')
df = df[df['Monthly Income'].notna()]

# Convertir 'Attrition' a valores numéricos si es necesario
df['Attrition'] = df['Attrition'].apply(lambda x: 1 if x == 'Left' else 0)

# Asegurarse de que hay suficiente variación en 'Monthly Income'
if df['Monthly Income'].nunique() < 4:
    print("Warning: 'Monthly Income' has limited unique values, quantiles might not be meaningful.")
    # Considerar métodos alternativos de agrupación si la variación es baja
    bin_edges = [0, 5000, 10000, 15000, 20000, float('inf')]  # Ajustar según sea necesario
    income_groups = pd.cut(df['Monthly Income'], bins=bin_edges, labels=['0-5k', '5k-10k', '10k-15k', '15k-20k', '20k+'], include_lowest=True)
else:
    # Agrupaciones por cuantiles
    income_groups = pd.qcut(df['Monthly Income'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

# Gráfico de boxplot para tasa de deserción por ingreso mensual
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='Monthly Income', data=df)
plt.title('Tasa de Deserción por Ingreso Mensual')
plt.show()

# Calcular las tasas de deserción para diferentes grupos de ingresos
attrition_by_income = df.groupby(income_groups, observed=True)['Attrition'].mean()
if attrition_by_income.empty:
    print("Warning: No attrition rates calculated. Check data or grouping.")
else:
    print("Attrition Rate by Income Group:\n", attrition_by_income)

    # Crear gráfico de barras para mostrar las tasas de deserción por grupos de ingresos
    attrition_by_income.plot(kind='bar', figsize=(10, 6), color='skyblue')
    plt.title('Tasa de Deserción por Grupo de Ingresos Mensuales')
    plt.ylabel('Tasa de Deserción')
    plt.xlabel('Grupo de Ingresos Mensuales')
    plt.xticks(rotation=45)  # Rotar etiquetas del eje x para mejor legibilidad
    plt.grid(True)  # Agregar una cuadrícula para mejor visualización
    plt.show()
```
